```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from scipy.stats import laplace
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
class CPU:
    def __init__(self, clockSpeed, cpuMark, price):
        self.clockSpeed = clockSpeed
        self.cpuMark = cpuMark
        self.price = price
        
    def randDist(self):
        self.Dist=laplace.rvs(loc=self.cpuMark,scale=self.price,size=self.clockSpeed)    
    
    def probDF(self):
        self.probDF = laplace.pdf(self.Dist)
        sns.lineplot(x=self.Dist, y=self.probDF)
        plt.show()

    def cumulativeDF(self):
        self.cumulativeDF = laplace.cdf(self.Dist)
        sns.lineplot(x=self.Dist, y=self.cumulativeDF)
        plt.show()
        
```

```python
cpu_1 = CPU(36000,0.31,0.669)
```

```python
cpu_1.randDist()
ax=sns.distplot(cpu_1.Dist, color="blue")
```

```python
cpu_1.probDF()
```

```python
cpu_1.cumulativeDF()
```
